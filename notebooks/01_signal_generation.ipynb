# Detector Signal Generation

In this notebook, we simulate a simple particle detector waveform. 

Particle detectors often produce signals with characteristic pulse shapes when a particle interacts with the detector. These pulses are then read out electronically, which can introduce noise.  

For this simulation, we model the detector response as a **Gaussian pulse**, which is a good approximation for many detector readouts. Later, we will add realistic noise to these signals and apply both classical and machine learning-based noise reduction techniques.

**Parameters of the pulse:**
- **Amplitude (A):** peak height of the pulse
- **Center (t0):** time of the pulse peak
- **Width (sigma):** how wide the pulse is

We will visualize the clean signal before adding noise.


```python
import numpy as np
import matplotlib.pyplot as plt

# Signal parameters
A = 1.0       # amplitude
t0 = 5.0      # center of pulse (arbitrary units)
sigma = 0.5   # pulse width
t = np.linspace(0, 10, 1000)  # time axis

# Gaussian pulse formula
signal = A * np.exp(-(t - t0)**2 / (2 * sigma**2))

# Plot the clean signal
plt.figure(figsize=(8,4))
plt.plot(t, signal, label='Clean Signal', color='blue')
plt.xlabel('Time (arb. units)')
plt.ylabel('Amplitude')
plt.title('Simulated Detector Signal (Gaussian Pulse)')
plt.legend()
plt.grid(True)
plt.show()
```
